# Step 1 - Data Engineering

```python
# import hawaii_measurements
hm = pd.read_csv('hawaii_measurements.csv', encoding="utf-8")
hm.head()
hm.tail()
hm.info()
hm.shape
```

```python
# use pandas profiling to print an overview report
hm_profile = pandas_profiling.ProfileReport(hm)
hm_profile
```

```python
# filter to see NaN on prcp
null_prcp = hm[hm.prcp.isnull()]
null_prcp.head()
null_prcp.tail()
null_prcp.shape
```

```python
# drop null values of prcp in hm
hm_drop = hm.dropna(axis=0, how='any')
hm_drop.head()
hm_drop.tail()
hm_drop.info()
hm_drop.shape
```

# export to csv
hm_drop.to_csv('clean_measurements.csv', encoding="utf-8", index=False)

```python
# import hawaii_measurements
hs = pd.read_csv('hawaii_stations.csv', encoding="utf-8")
hs
hs.info()
hs.shape
```

```python
# merge hm & hs on station
merge1 = pd.merge(hm, hs, on='station')
merge1.head()
merge1.tail()
merge1.info()
merge1.shape
# pd.merge(df_new, df_n, on='subject_id')
```

```python
# sort merge 1 by date
sorted_merge1 = merge1.sort_values(['date'], ascending=False)
sorted_merge1.head()
```

```python
# use pandas profiling to print an overview report
merge1_profile = pandas_profiling.ProfileReport(merge1)
merge1_profile
```

```python
# drop NaN values
merge1_drop = merge1.dropna(axis=0, how='any')
merge1_drop.head()
merge1_drop.tail()
merge1_drop.info()
merge1_drop.shape
```

**tobs = Time of observation Bias**

```python
# use pandas profiling to analyze merge1_drop
merge1_profile_drop = pandas_profiling.ProfileReport(merge1_drop)
merge1_profile_drop
```

# export cleaned df to csv
merge1_drop.to_csv('clean_hawaii_merge.csv', encoding='utf-8', index=False)
