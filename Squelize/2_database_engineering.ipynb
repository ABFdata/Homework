```python
# import cleaned df
# first sheet is a merge of both measurements & stations
ch_df = pd.read_csv('clean_hawaii_merge.csv', encoding='utf-8')
ch_df.head()
ch_df.tail()
ch_df.info()
ch_df.shape
```

```python
# import cleaned measurements
# this sheet is a cleaned sheet of measurements. null values are gone
m_df = pd.read_csv('clean_measurements.csv', encoding='utf-8')
m_df.head()
m_df.tail()
m_df.info()
m_df.shape
```

```python
# import the stations table
hs = pd.read_csv('hawaii_stations.csv', encoding="utf-8")
hs
hs.info()
hs.shape
```

```python
# declarative_base
Base = declarative_base()
```

```python
# Create Measurement Class
# these classes are empty classes with no data
class Measurement(Base):
    __tablename__ = 'measurement'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    date = Column(Text(64))
    prcp = Column(Float(64))
    tobs = Column(Float(64))
```

```python
# Create Station Class
# these classes are empty classes with no data
class Station(Base):
    __tablename__='station'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    name = Column(String(255))
    latitude = Column(Float(64))
    longitude = Column(Float(64))
    elevation = Column(Float(64))
```

```python
# Use the engine and connection string to create a database 
# called hawaii.sqlite

engine = create_engine("sqlite:///hawaii.sqlite", echo=False)
conn = engine.connect()
```

```python
# create a "metadata" layer that abstracts our SQL Database
Base.metadata.create_all(engine)
```

```python
# create a session object to connect to DB
session = Session(bind=engine)
```

```python
# add cleaned measurements and stations csv data 

# m_df - measurements table - appends to hawaii.sqlite
m_df.to_sql('measurement', conn, if_exists='append', index=False)

# hs - stations table - appends to hawaii.sqlite
hs.to_sql('station', conn, if_exists='append', index=False)
```

```python
# read back measurement able to see if they we're appended to hawaii.sqlite

data_measurement = pd.read_sql("SELECT * FROM measurement", conn)
# drop column id in data_measurement Note: axis=1 denotes that we are referring to a column, not a row
measurement_drop_index = data_measurement.drop('id', axis=1)

# drop duplicates
measurement_clean = measurement_drop_index.drop_duplicates()
measurement_clean.head()
measurement_clean.tail()
measurement_clean.shape
```

```python
# read back station table to see if they we're appended to hawaii.sqlite

data_station = pd.read_sql("SELECT * FROM station", conn)

# drop column id in data_station Note: axis=1 denotes that we are referring to a column, not a row
station_drop_index = data_station.drop('id', axis=1)

# drop duplicates
station_clean = station_drop_index.drop_duplicates()
station_clean.head()
station_clean.tail()
station_clean.shape
```

```python
# Merge the columns of measurement and station DF that were appeneded to hawaii.sqlite
combinded_data = pd.merge(measurement_clean, station_clean, on="station", how="inner")
combinded_data.head()
combinded_data.shape
```
